```python
# === Light EDA for Cleaned Dataset 1 ===
import pandas as pd

# --- Load cleaned dataset (update if needed) ---
df = pd.read_csv("cleaned_dataset1.csv")

print("âœ… Dataset loaded")
print("="*50)

# --- Shape and column info ---
print("Shape:", df.shape)
print("\nColumns & dtypes:")
print(df.dtypes)
print("="*50)

# --- First few rows ---
print("\nHead of dataset:")
print(df.head())
print("="*50)

# --- Missing values ---
print("\nMissing values per column:")
print(df.isnull().sum())
print("="*50)

# --- Descriptive statistics (numerical) ---
print("\nNumerical summary:")
print(df.describe())
print("="*50)

# --- Descriptive statistics (categorical) ---
print("\nCategorical summary:")
print(df.describe(include=["category", "object"]))
print("="*50)

# --- Value counts for key variables ---
key_vars = [
    "income_level", "education_level", "rural_or_urban", "gender",
    "smoking_status", "diet_score", "hypertension", "diabetes",
    "obesity", "alcohol_consumption", "heart_attack"
]
for col in key_vars:
    if col in df.columns:
        print(f"\n--- {col} ---")
        print(df[col].value_counts(normalize=True).round(3) * 100)
```

```python
# Save summaries for sharing
df.describe().to_csv("numerical_summary.csv")
df.describe(include=['category', 'object']).to_csv("categorical_summary.csv")
```
