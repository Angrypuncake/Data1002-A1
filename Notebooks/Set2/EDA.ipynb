```python
import pandas as pd       
import numpy as np        
import matplotlib.pyplot as plt  
from pathlib import Path 

DATA_PATH = Path('heart_attack_risk_India_cleaned.csv')  

pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 160)

df = pd.read_csv(DATA_PATH)

print('Loaded shape:', df.shape)

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.barplot(x="heart_attack_risk", y="diet_category", data=df, palette="Set2", estimator="mean", ci=None)

plt.title("Average Diet Category by Heart Attack Risk")
plt.xlabel("Heart Attack Risk")
plt.ylabel("Average Diet Score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Find average heart disease rate in each income group
cvd_rates = df.groupby('income_group')['heart_attack_risk'].mean()

# Make a bar chart
cvd_rates.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Heart Attack Rates by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Proportion with Heart Attack Risk')
plt.ylim(0, 1)  # makes y-axis go from 0 to 1 (0% to 100%)
plt.show()
```
